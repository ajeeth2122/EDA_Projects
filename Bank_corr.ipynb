```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Asus\Downloads\bank.csv")
```

```python
df
```

```python
for c in df.columns:
    missing = np.mean(df[c].isnull())
    print('{}-{}%'.format(c,missing))
```

```python
for c in df.columns:
    if df[c].dtype == 'object':
        df[c]= df[c].astype("category")
        df[c]=df[c].cat.codes
```

```python
for i in df.columns:
    df[i] = df[i].astype("int64")
```

```python
df.info()
```

```python
cor = df.corr()
```

```python
plt.scatter(x=df['month'] , y=df['balance'])
```

```python
plt.figure(figsize=(12,2))
sns.regplot(x=df['month'] , y=df['balance'])
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(cor,annot = True)
```
